# The metallicity effect of the NIKE diagram
It seems that the edge has an effect of [Fe/H], though it is hard to spot. Sanjib mentions there is a metallicity effect from the isochrones.

Why is it not a data artifact?
Maybe arise from the unenven number of samples - hardly true after plotting according to histogram.

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111)
idx = (yu["Phase"]==2)
h = ax.hist(yu["[Fe/H]"], bins=np.arange(-2.5,1.0,0.1))

mean=h[1][np.where(h[0]==h[0].max())[0][0]]+0.05
ax.axvline(mean, linestyle="--", color="black") 
ax.text(mean, np.mean(ax.get_ylim()), "{:0.2f}".format(mean),  color="black")
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111)
ax.axis([10, 200, 2.3, 5.0])
ax.set_xscale("log")

feh_limits = [-2.0, -0.05, 1.0] 
colors = ["blue", "red"] #sns.diverging_palette(220, 20, n=len(feh_ll))
labels = ["-2.0<[Fe/H]<-0.05", "-0.05<[Fe/H]<1.0"]
for igroup in range(len(feh_limits)-1):
    idx = (yu["[Fe/H]"]>=feh_limits[igroup]) & (yu["[Fe/H]"]<feh_limits[igroup+1]) & (yu["Phase"]==2)
    x, y = yu["numax"][idx], yu["numax"][idx]**0.75/yu["Delnu"][idx]
    ax.scatter(x, y, color=colors[igroup], s=2, label=labels[igroup])

ax.set_title("Purely observations")
ax.legend()

# idx = (yu["[Fe/H]"]>=-1.0) & (yu["[Fe/H]"]<1.0) & (yu["Phase"]==2)
# x, y = yu["numax"][idx], yu["numax"][idx]**0.75/yu["Delnu"][idx]
# c = ax.scatter(x, y, c=yu["[Fe/H]"][idx], s=2, cmap="seismic")
# plt.colorbar(c, ax=ax, orientation='vertical')
```

# quantify the dege
## on the nike diagram

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111)
ax.axis([10, 200, 2.3, 5.0])
ax.set_xscale("log")

idx = (yu["Phase"]==2)
xobs, yobs = yu["numax"][idx], yu["numax"][idx]**0.75/yu["Delnu"][idx]
Nobs = xobs.shape[0]
# plot observations
ax.plot(xobs, yobs, "r.", markersize=2)
ax.set_title("Purely observations")
ax.grid(which="both")
ax.set_xlabel("numax")
ax.set_ylabel("numax^0.75/dnu")

# plot edge defining points
points=[(32, 2.9), (30.7, 3.0), (35.5, 3.35), (40, 3.5), (60, 3.75), (100, 3.85), (112, 3.86)]
Npoint = len(points)
xpoint, ypoint = [points[i][0] for i in range(Npoint)], [points[i][1] for i in range(Npoint)]
ax.plot(xpoint, ypoint, "k.", markersize=10)

# cubic spline interpolation
import scipy
tck = scipy.interpolate.splrep(ypoint, np.log10(xpoint), s=0)
yedge = np.arange(2.5, 4.0, 0.001)
xedge = 10.0**scipy.interpolate.splev(yedge, tck, der=0)
Nedge = xedge.shape[0]
ax.plot(xedge, yedge, "k-") 

# f = interpolate.interp1d(y, np.log10(x), kind="cubic")
# yedge = np.arange(2.9, 3.8, 0.01)
# xedge = 10.0**f(yedge)
# ax.plot(xedge, yedge, "b-") 


# left or right?
idx_left = xobs<=10.0**scipy.interpolate.splev(yobs, tck, der=0)
idx_right = xobs>10.0**scipy.interpolate.splev(yobs, tck, der=0)
idx_mrich = yu["[Fe/H]"][idx]>=0
idx_mpoor = yu["[Fe/H]"][idx]<0

# shortest distance distribution
X = np.zeros((Nobs+Nedge, 2))
X[0:Nobs,0], X[0:Nobs,1] = xobs, yobs
X[Nobs:,0], X[Nobs:,1] = xedge, yedge
Y = scipy.spatial.distance.pdist(X)
sdist=Y[0:Nobs]
sdist[idx_right]=-sdist[idx_right]
np.save("sample/edge_nike", X[Nobs:,])


fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111)
ax.set_xlim(-15, 20)
h=ax.hist(sdist[idx_mpoor], color="red", histtype="step", label="Metal poor",
    bins=np.linspace(sdist.min(), sdist.max(), 1500), zorder=0)
h=ax.hist(sdist[idx_mrich], color="red", linestyle="--", histtype="step", label="Metal rich",
    bins=np.linspace(sdist.min(), sdist.max(), 1500), zorder=0)
ax.set_xlabel("Distances to the edge")
# h=ax.hist(sdist[idx_right], color="orange", label="Points at right of the curve",
#      bins=np.linspace(sdist.min(), sdist.max(), 100), zorder=1)
ax.legend()
```
