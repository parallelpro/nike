```python
numax_solar, dnu_solar, teff_solar = 3090, 135.1, 5777

def get_emass_eradius(numax, dnu, teff, enumax, ednu, eteff, numax_solar, dnu_solar, teff_solar):
    emass = ((3*(numax/numax_solar)**2. * (dnu/dnu_solar)**-4 * (teff/teff_solar)**1.5)**2. * (enumax/numax_solar)**2.
    + ((numax/numax_solar)**3. * (-4)*(dnu/dnu_solar)**-5 * (teff/teff_solar)**1.5)**2. * (ednu/dnu_solar)**2.0
    + ((numax/numax_solar)**3. * (dnu/dnu_solar)**-4 * 1.5*(teff/teff_solar)**0.5)**2. * (eteff/teff_solar)**2.0 )**0.5
    eradius = (( (dnu/dnu_solar)**-2 * (teff/teff_solar)**0.5 )**2. * (enumax/numax_solar)**2.
    + ((numax/numax_solar)**1. * (-2)*(dnu/dnu_solar)**-3 * (teff/teff_solar)**0.5)**2. * (ednu/dnu_solar)**2.0
    + ((numax/numax_solar)**1. * (dnu/dnu_solar)**-2 * 0.5*(teff/teff_solar)**-0.5)**2. * (eteff/teff_solar)**2.0 )**0.5
    return emass, eradius

def get_scaling_mass_radius(numax, dnu, teff, numax_solar, dnu_solar, teff_solar):
    mass = (numax/numax_solar)**3. * (dnu/dnu_solar)**-4 * (teff/teff_solar)**1.5
    radius = (numax/numax_solar)**1. * (dnu/dnu_solar)**-2 * (teff/teff_solar)**0.5 
    return mass, radius



# set up data
idx = (yu["Phase"]==2) #& (np.array(yu["numax"]**0.75/yu["Delnu"])>3.0)
# edges = np.load(rootpath+"sample/obs/tnu_edge_samples.npy")
# tck = np.load(rootpath+"sample/obs/spline_tck.npy", allow_pickle=True)
# dist = distance_to_edge(yu["numax"], yu["numax"]**0.75/yu["Delnu"], 
#                         edges[:,0], edges[:,1], tck, diagram="nike", distance="horizontal")

# idx = (yu["Phase"]==2) & (dist>-2)

numax_obs, nike_obs, dnu_obs = np.asarray(yu["numax"][idx]), np.asarray(yu["numax"][idx]**0.75/yu["Delnu"][idx]), np.asarray(yu["Delnu"][idx])
Nobs = numax_obs.shape[0]
feh_obs = np.asarray(yu["[Fe/H]"][idx])


e_numax_obs = np.asarray(yu["e_numax"][idx])
a, b = yu["numax"][idx], yu["Delnu"][idx]
e_nike_obs = np.asarray(((0.75*a**-0.25*b**-1*yu["e_numax"][idx])**2.0 + (a**0.75*b**-2*yu["e_Delnu"][idx])**2.0)**0.5)
e_dnu_obs = np.array(yu["e_Delnu"][idx])
e_feh_obs = np.asarray(yu["e_[Fe/H]"][idx])
teff_obs = np.asarray(yu["Teff"][idx])
e_teff_obs = np.asarray(yu["e_Teff"][idx])


from fdnu import asfgrid
s=asfgrid.Seism(datadir=rootpath+"fdnu/")
fnumax = 1
evstate = np.zeros(len(feh_obs))+2
mass_obs, radius_obs = s.get_mass_radius(evstate, feh_obs, teff_obs, dnu_obs, 
                           numax_obs, fill_value=None, isfeh=True)

e_mass_obs, e_radius_obs = get_emass_eradius(numax_obs, dnu_obs, teff_obs, 
                                             e_numax_obs, e_dnu_obs, e_teff_obs, 
                                             numax_solar, dnu_solar, teff_solar)

mass_obs_nc, radius_obs_nc = get_scaling_mass_radius(numax_obs, dnu_obs, teff_obs,
                                             numax_solar, dnu_solar, teff_solar)
e_mass_obs_nc, e_radius_obs_nc = e_mass_obs, e_radius_obs



idx = np.isfinite(mass_obs) & np.isfinite(radius_obs)

data = {"dnu":dnu_obs[idx], "numax":numax_obs[idx], "feh": feh_obs[idx], "mass":mass_obs[idx], 
        "radius":radius_obs[idx], "mass_nc": mass_obs_nc[idx], "radius_nc":radius_obs_nc[idx], 
        "nike":nike_obs[idx], "teff":teff_obs[idx],
       "e_dnu":e_dnu_obs[idx], "e_numax":e_numax_obs[idx], 
        "e_feh": e_feh_obs[idx], "e_mass":e_mass_obs[idx], 
        "e_radius":e_radius_obs[idx],
       "e_mass_nc": e_mass_obs_nc[idx], "e_radius_nc":e_radius_obs_nc[idx], 
        "e_nike":e_nike_obs[idx], "e_teff":e_teff_obs[idx]}

filepath = rootpath+"sample/yu/"
np.save(filepath+"yu18", data)

filepath = rootpath+"sample/yu_nc/"
np.save(filepath+"yu18", data)
```

```python
def edge_analysis(xdata, ydata, ydatat, filepath, points, tp, title, ifplot=True):
    # plot two diagrams
    fig = plt.figure(figsize=(12,12))
    axes = fig.subplots(nrows=2,ncols=1)
#     axes[0].axis([10, 200, 2.3, 5.0])
#     axes[0].set_xscale("log")
    for i in range(2):
#         axes[i].axis([10, 200, 2.3, 10.0])
        axes[i].axis([10, 200, 2.7, 4.2])
        axes[i].set_xscale("log")

        axes[i].set_title(title)
        axes[i].grid(which="both")
        axes[i].set_xlabel("numax")
        axes[i].set_ylabel("dnu")

    # cubic spline interpolation
    # points
    tcks = []
    ymin = min(np.array([points[0][i][1] for i in range(len(points[0]))]).min(), np.array([points[1][i][1] for i in range(len(points[1]))]).min())
    ymax = max(np.array([points[0][i][1] for i in range(len(points[0]))]).max(), np.array([points[1][i][1] for i in range(len(points[1]))]).max())
    yedge = np.arange(ymin, ymax, 0.0001)
    xedge = np.zeros(len(yedge))
    idx = [yedge<tp[1], yedge>=tp[1]]
    for j, point in enumerate(points):
        Npoint = len(point)
        xpoint, ypoint = np.array([point[i][0] for i in range(Npoint)]), np.array([point[i][1] for i in range(Npoint)])
        import scipy.interpolate
        tck = scipy.interpolate.splrep(np.log10(ypoint), np.log10(xpoint), s=0)
        tcks.append(tck)

        # plot knots and edge 
#         axes[0].plot(xpoint, xpoint**0.75/ypoint, "gx", markersize=10)
#         axes[1].plot(xpoint, xpoint**0.75/ypoint, "gx", markersize=10)
        axes[0].plot(xpoint, ypoint, "gx", markersize=10,markeredgewidth=2,zorder=100)
        axes[1].plot(xpoint, ypoint, "gx", markersize=10,markeredgewidth=2,zorder=100)
        xedge[idx[j]] = 10.0**scipy.interpolate.splev(np.log10(yedge[idx[j]]), tck, der=0)
    
    yedge = xedge**0.75/yedge
    axes[0].plot(xedge, xedge**0.75/yedge, "k.", ms=1, zorder=10) 
    axes[1].plot(xedge, xedge**0.75/yedge, "k.", ms=1, zorder=10) 
    
    # save spline coefficients
    np.save(filepath+"nike_spline_tck", [tcks, tp])
    

    # save edges
    np.save(filepath+"nike_edge_samples", np.array([xedge, xedge**0.75/yedge]).T)
    np.save(filepath+"tnu_edge_samples", np.array([xedge, yedge]).T)


    dx, numaxx, dnux = distance_to_edge(xdata, ydatat, xedge, yedge, tcks, tp, diagram="tnu", distance="horizontal")
    dy, numaxy, dnuy = distance_to_edge(xdata, ydatat, xedge, yedge, tcks, tp, diagram="tnu", distance="vertical")
    axes[0].plot(xdata, ydata, ".", color="gray",markersize=5)
    axes[0].plot(numaxx, numaxx**0.75/dnux, "r.", markersize=2, label="horizontal data")
    axes[1].plot(xdata, ydata, ".", color="gray", markersize=5)
    axes[1].plot(numaxy, numaxy**0.75/dnuy, "b.", markersize=2, label="vertical data")
    axes[0].legend()
    axes[1].legend()
    
#     # # now plot histograms
#     fig = plt.figure(figsize=(12,12))
#     axes = fig.subplots(nrows=2,ncols=1)
#     axes[0].set_xlim(-20, 82)
#     axes[1].set_xlim(-2, 4)
#     bins0 = np.linspace(-20, 82, 100)
#     bins1 = np.linspace(-2, 12, 300)
#     # bins0 = np.linspace(distn.min(), distn.max(), 1500)
#     # bins1 = np.linspace(distt.min(), distt.max(), 1500)
#     h=axes[0].hist(dx, histtype="step", color="red",
#         bins=bins0, zorder=0)
#     h=axes[1].hist(dy, histtype="step", color="red",
#         bins=bins1, zorder=0)
#     axes[0].set_xlabel("Horizontal distance")
#     axes[1].set_xlabel("Vertical distance")
#     axes[0].set_ylabel("Ncount")
#     axes[1].set_ylabel("Ncount")
#     axes[0].grid(which="both")
#     axes[1].grid(which="both")
#     axes[0].set_title(title)
#     axes[1].set_title(title)
#     plt.show()
    if ifplot: plt.show()
    return
```

```python
numax_solar, dnu_solar, teff_solar = 3090, 135.1, 5777

def get_emass_eradius(numax, dnu, teff, enumax, ednu, eteff, numax_solar, dnu_solar, teff_solar):
    emass = ((3*(numax/numax_solar)**2. * (dnu/dnu_solar)**-4 * (teff/teff_solar)**1.5)**2. * (enumax/numax_solar)**2.
    + ((numax/numax_solar)**3. * (-4)*(dnu/dnu_solar)**-5 * (teff/teff_solar)**1.5)**2. * (ednu/dnu_solar)**2.0
    + ((numax/numax_solar)**3. * (dnu/dnu_solar)**-4 * 1.5*(teff/teff_solar)**0.5)**2. * (eteff/teff_solar)**2.0 )**0.5
    eradius = (( (dnu/dnu_solar)**-2 * (teff/teff_solar)**0.5 )**2. * (enumax/numax_solar)**2.
    + ((numax/numax_solar)**1. * (-2)*(dnu/dnu_solar)**-3 * (teff/teff_solar)**0.5)**2. * (ednu/dnu_solar)**2.0
    + ((numax/numax_solar)**1. * (dnu/dnu_solar)**-2 * 0.5*(teff/teff_solar)**-0.5)**2. * (eteff/teff_solar)**2.0 )**0.5
    return emass, eradius

def get_scaling_mass_radius(numax, dnu, teff, numax_solar, dnu_solar, teff_solar):
    mass = (numax/numax_solar)**3. * (dnu/dnu_solar)**-4 * (teff/teff_solar)**1.5
    radius = (numax/numax_solar)**1. * (dnu/dnu_solar)**-2 * (teff/teff_solar)**0.5 
    return mass, radius


# set up data
idx = (apk["Phase"]==2)

numax_obs, nike_obs, dnu_obs = np.asarray(apk["numax"][idx]), np.asarray(apk["numax"][idx]**0.75/apk["Delnu"][idx]), np.asarray(apk["Delnu"][idx])
Nobs = numax_obs.shape[0]
feh_obs = np.asarray(apk["FeH"][idx])
teff_obs = np.asarray(apk["Teff_2"][idx])
e_teff_obs = np.asarray(apk["e_Teff_2"][idx])

from fdnu import asfgrid
s=asfgrid.Seism(datadir=rootpath+"fdnu/")
fnumax = 1
# mh = apk["FeH"][idx] + np.log10(10.0**apk["AFe"][idx]*0.694 + 0.306)
mh = apk["FeH"][idx]
evstate = np.zeros(len(mh))+2
km, kr = s.get_mass_radius(evstate, mh, apk["Teff_2"][idx], apk["Delnu"][idx], apk["numax"][idx], fill_value=None, isfeh=True)

mass_obs = km
radius_obs = kr

e_numax_obs = np.asarray(apk["e_numax"][idx])
a, b = apk["numax"][idx], apk["Delnu"][idx]
e_nike_obs = np.asarray(((0.75*a**-0.25*b**-1*apk["e_numax"][idx])**2.0 + (a**0.75*b**-2*apk["e_Delnu"][idx])**2.0)**0.5)
e_dnu_obs = np.array(apk["e_Delnu"][idx])
e_feh_obs = np.asarray(apk["e_[Fe/H]"][idx])
e_mass_obs, e_radius_obs = get_emass_eradius(numax_obs, dnu_obs, teff_obs, 
                                             e_numax_obs, e_dnu_obs, e_teff_obs, 
                                             numax_solar, dnu_solar, teff_solar)

mass_obs_nc, radius_obs_nc = get_scaling_mass_radius(numax_obs, dnu_obs, teff_obs,
                                             numax_solar, dnu_solar, teff_solar)
e_mass_obs_nc, e_radius_obs_nc = e_mass_obs, e_radius_obs

idx = np.isfinite(mass_obs) & np.isfinite(radius_obs)
data = {"dnu":dnu_obs[idx], "numax":numax_obs[idx], "feh": feh_obs[idx], "mass":mass_obs[idx], "radius":radius_obs[idx],
       "mass_nc": mass_obs_nc[idx], "radius_nc":radius_obs_nc[idx], "nike":nike_obs[idx], "teff":teff_obs[idx],
       "e_dnu":e_dnu_obs[idx], "e_numax":e_numax_obs[idx], "e_feh": e_feh_obs[idx], "e_mass":e_mass_obs[idx], "e_radius":e_radius_obs[idx],
       "e_mass_nc": e_mass_obs_nc[idx], "e_radius_nc":e_radius_obs_nc[idx], "e_nike":e_nike_obs[idx], "e_teff":e_teff_obs[idx]}

filepath = rootpath+"sample/yu/"
np.save(filepath+"apk18", data)

filepath = rootpath+"sample/yu_nc/"
np.save(filepath+"apk18", data)
```

```python
# set up data
mist=np.load(rootpath+"sample/kepler_galaxia_mist_uncofeh.npy")
Nstar = mist.shape[0]

idx = (mist["phase"]>=3) & (mist["numax"]**0.75/mist["dnu"]>2.8) #(np.isfinite(mist["numax"])) & (np.isfinite(mist["dnu"])) & (np.isfinite(synp["feh"])) & (np.array(synp["numax"]**0.75/synp["dnu"])>3.0)
numax_mst= np.array(mist["numax"][idx])
nike_mst = np.array(mist["numax"][idx]**0.75/mist["dnu"][idx])
dnu_mst = np.array(mist["dnu"][idx])
# Nmst = xmst.shape[0]
mass_mst = np.array(mist["mass"][idx])
radius_mst =  np.array(mist["radius"][idx])
feh_mst = mist["feh"][idx]
teff_mst = np.asarray(mist["Teff"][idx])
dnu_nc_mst = np.array(mist["dnu_scaling"][idx])
nike_nc_mst = np.array(mist["numax"][idx]**0.75/mist["dnu_scaling"][idx])

data = {"dnu":dnu_mst, "numax":numax_mst, "nike":nike_mst, "feh": feh_mst, "mass":mass_mst, "radius":radius_mst,
       "dnu_nc": dnu_nc_mst, "nike_nc":nike_nc_mst}

filepath = rootpath+"sample/heb/mist/"
np.save(filepath+"mist", data)
filepath = rootpath+"sample/heb/mist_nc/"
np.save(filepath+"mist", data)

# filepath = rootpath+"sample/mist_nc/"
# np.save(filepath+"padova", data)
```

# Galaxia with Kepler uncertainties


```python
# set up data
import ebf
synp=ebf.read(rootpath+"sample/kepler_galaxia_mrtd5.ebf")
Nstar = synp["alpha"].shape[0]
# factor = int(Nstar/16000)
# idx = np.arange(0,int(Nstar/factor))*factor
# for key in synp.keys():
#     synp[key] = synp[key][idx]


# # corrected dnu and numax
idx = (synp["evstate"]==2) & (np.isfinite(synp["numax"])) & (np.isfinite(synp["dnu"])) & (np.isfinite(synp["feh"])) & (np.array(synp["numax"]**0.75/synp["dnu"])>3.0)
numax, nike, dnu = np.array(synp["numax"][idx]), np.array(synp["numax"][idx]**0.75/synp["dnu"][idx]), np.array(synp["dnu"][idx])
dnu_nc = np.array(synp["dnu_scaling"][idx])
nike_nc = np.array(numax**0.75/dnu_nc)
mass, radius =  np.array(synp["mact"][idx]), np.array(synp["radius"][idx])
feh = synp["feh"][idx]

Npdv = numax.shape[0]

data = {"dnu":dnu, "numax":numax, "nike":nike, "feh": feh, "mass":mass, "radius":radius,
       "dnu_nc": dnu_nc, "nike_nc":nike_nc}

filepath = rootpath+"sample/padova_err/"
np.save(filepath+"padova", data)
```
