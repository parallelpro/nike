# Mass bins

```python
scatter, escatter, scalar, escalar = [np.zeros(5) for i in range(4)]
mass = np.arange(0.9, 1.9, 0.2)
Nbins = len(mass)
for iname in ["numax", "dnu", "mass", "radius"]:
    for i in range(Nbins):
        tfilepath = rootpath+"sample/sharpness/sharma16/"+iname+"_mass/perturb/"+str(i)+"/data.npy"
        scatter[i] = np.load(tfilepath, allow_pickle=True).tolist()["para_fit"][-1]*100
        escatter[i] = np.load(tfilepath, allow_pickle=True).tolist()["e_para_fit"][-1]*100
        
        tfilepath = rootpath+"sample/sharpness/sharma16/"+iname+"_mass/rescale/"+str(i)+"/data.npy"
        scalar[i] = np.load(tfilepath, allow_pickle=True).tolist()["para_fit"][-1]*100
        escalar[i] = np.load(tfilepath, allow_pickle=True).tolist()["e_para_fit"][-1]*100
        
    fig = plt.figure(figsize=(7,5))
    ax = fig.add_subplot(111)
    ax.errorbar(mass, scatter, color="C0", yerr=escatter, fmt=".", capsize=2)
    axt = ax.twinx()
    axt.errorbar(mass, scalar, color="C1", yerr=escalar, fmt=".", capsize=2)
    axt.set_ylabel("Rescale factor [%]")
    ax.set_ylabel("Upper limits of intrinsic scatters [%]")
```
