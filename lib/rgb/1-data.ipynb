```python
def data_plot(data, size, title):
    fig, axes = plt.subplots(figsize=(10,10), nrows=2,ncols=2, squeeze=False)
    axes = axes.reshape(-1)
    import matplotlib.colors as colors
    n = colors.Normalize(vmin=-0.5, vmax=0.5)
    axes[0].scatter(data['teff'], data['numax'], c=data['feh'], marker='.', s=size, cmap='jet',norm=n)
    axes[1].scatter(data['teff'], data['dnu'], c=data['feh'], marker='.', s=size, cmap='jet',norm=n)
    axes[2].scatter(data['teff'], data['radius'], c=data['feh'], marker='.', s=size, cmap='jet',norm=n)
    axes[3].scatter(data['mass'], data['radius'], c=data['feh'], marker='.', s=size, cmap='jet',norm=n)

    axes[0].axis([5200, 4000, 100, 1])
    axes[0].set_xlabel('Teff')
    axes[0].set_ylabel('numax')
    axes[1].axis([5200, 4000, 10, 0.1])
    axes[1].set_xlabel('Teff')
    axes[1].set_ylabel('dnu')
    axes[2].axis([5200, 4000, 2., 20.])
    axes[2].set_xlabel('Teff')
    axes[2].set_ylabel('radius')
    axes[3].axis([0.2, 3.0, 2., 20.])
    axes[3].set_xlabel('mass')
    axes[3].set_ylabel('radius')
    # axes.set_yscale('log')
    
    axes[0].text(0.1, 0.9, title, transform=axes[0].transAxes)
    return

data_plot(apk, 1.0, 'Kepler')
```

```python
def best_clustered_line(kini, bini, xdata, ydata, bins, dist_limits, fittype='vertical'):
    
    def scatterfunc(theta):
        k, b = theta[0], theta[1]
        if fittype=='vertical':
            dist = ydata - (k*xdata+b)
        else:
            dist = xdata - (ydata-b)/k
        idx = np.abs(dist)<dist_limits

        scatter = np.std(dist[idx])
        return scatter

    para_guess = np.array([kini, bini])
    bounds = [[t*0.2, t*2.0] for t in para_guess]
    res = optimize.minimize(scatterfunc, para_guess, bounds=bounds)
    kbest, bbest = res.x
    
    return kbest, bbest

def workflow(x, y, p1, p2, title, filepath):
    fig = plt.figure(figsize=(10,5))
    ax=fig.add_subplot(121)
    Nstar = len(x)
    idx = np.array(np.arange(0,Nstar)/Nstar * (len(x)), dtype=int)
    c=ax.scatter(x[idx], y[idx], c='gray', marker='.', s=1)
    ax.axis([5500, 4000, 120, 10])
    # ax.set_yscale('log')
    
    k = (p2[1]-p1[1])/(p2[0]-p1[0])
    b = p2[1]-k*p2[0]
    tx = np.linspace(4000, 5500, 100)
    ty = tx*k + b
    ax.plot(tx, ty, 'r-')
    ax.grid(True)
    
    ax.set_xlabel('Teff')
    ax.set_ylabel('numax')
    ax.text(0.1, 0.9, title, color='r', transform=ax.transAxes)
    
    ax2 = fig.add_subplot(122)

    bins = np.arange(-30, 30, 2.0)
    dist_limits = 30.0
    kbest, bbest = best_clustered_line(k, b, x[idx], y[idx], bins, dist_limits, 'vertical')
    vd = y[idx] - (kbest*x[idx]+bbest)
    idx = np.abs(vd)<dist_limits
    ax.plot(tx, tx*kbest+bbest, 'b-')

    # vertical
    histy, bins, _ = ax2.hist(vd, histtype='step', bins=bins)
    histx = (bins[:-1]+bins[1:])/2.
    fitfunc = lambda p, x: p[0]*np.exp(-(x-p[1])**2.0/(2*p[2]**2.0)) + p[3]
    errfunc = lambda p, x, y: fitfunc(p, x) - y
    p0 = [np.max(histy), 0., 1.0, np.median(histy)] # Initial guess for the parameters
    p1, success = optimize.leastsq(errfunc, p0[:], args=(histx, histy))
    fitx = np.linspace(histx.min(), histx.max(), 100)
    ax2.plot(fitx, fitfunc(p1, fitx), "k--")
    ax2.text(0.65, 0.9, 'Scatter = {:0.2f}'.format(p1[2]), color='r', transform=ax2.transAxes)

    ax2.set_xlabel('Vertical distance')
    
    plt.show()
    np.save(filepath, np.array([kbest, bbest]))


filepath = rootpath+'sample/rgb/yu/numax_bump.npy'
workflow(apk['teff'], apk['numax'], [4580, 58], [4800, 35], 'Kepler', filepath)
# workflow(apk['teff'], apk['numax'], [4580, 60], [4800, 35], rootpath+'sample/rgb/yu_nc/numax_bump.npy')

filepath = rootpath+'sample/rgb/padova/numax_bump.npy'
workflow(padova['teff'], padova['numax'], [4400, 58], [4800, 21], 'Galaxia+PARSEC', filepath)
# workflow(padova['teff'], padova['numax'], [4400, 58], [4800, 21], rootpath+'sample/rgb/padova_nc/numax_bump.npy')

filepath = rootpath+'sample/rgb/mist/numax_bump.npy'
workflow(mist['teff'], mist['numax'], [4400, 54], [4800, 21], 'Galaxia+MIST', filepath)
# workflow(mist['teff'], mist['numax'], [4400, 58], [4800, 21], rootpath+'sample/rgb/mist_nc/numax_bump.npy')
```
